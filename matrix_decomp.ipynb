```python
import sympy as sp
```

```python
A = sp.Matrix(
    [
        [0, sp.Rational(1, 2), sp.Rational(1, 4), sp.Rational(1, 4)],
        [sp.Rational(1, 4), 0, sp.Rational(1, 2), sp.Rational(1, 4)],
        [sp.Rational(1, 3), sp.Rational(1, 3), 0, sp.Rational(1, 3)],
        [sp.Rational(1, 5), sp.Rational(2, 5), sp.Rational(2, 5), 0],
    ]
)

display(A)
```

```python
P, D = A.diagonalize()
```

```python
display(P)
```

```python
print('Abs D')
display(sp.Abs(D.evalf()))
```

```python
D_power = D.evalf()

for i in range(12):
    D_power = D_power ** 2

display(sp.cancel(D_power))

display((P.evalf() @ sp.Abs(D_power) @ P.evalf().inv()).evalf())
```

```python
n = sp.symbols('n')
```

```python
sp.simplify((P.evalf() @ sp.Abs(D.evalf()) ** n @ P.evalf().inv()).evalf())
```
